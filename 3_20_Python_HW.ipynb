# Part B: Saving, with a raise

```python
annual_salary = float(input('Enter your annual salary: '))
portion_saved = float(input('Enter the percent of your salary to save, as a decimal: '))
total_cost = float(input('Enter the cost of your dream home: '))
semi_annual_raise = float(input('Enter the semi-annual raise, as a decimal: '))

current_savings = 0
month = 0
r = 0.04
portion_down_payment = 0.25

while True:
    monthly_salary = (annual_salary / 12) * (1 + semi_annual_raise) ** (month // 6)
    current_savings = current_savings + monthly_salary * portion_saved + current_savings * r/12
    month = month + 1
    if current_savings >= total_cost * portion_down_payment:
        print('Number of months:', month)
        break
```

## 1. lyrics_to_frequencies(lyrics)

image.png

```python
def lyrics_to_frequencies(lyrics):
    lyrics_dict = {}

    for lyric in lyrics:
        lyrics_dict[lyric] = lyrics.count(lyric)

    return lyrics_dict
```

## 2. most_common_words(lyrics_dict)

image.png

```python
def most_common_words(lyrics_dict):

    values = lyrics_dict.values()
    highest_freq = max(values)
    words_list = []

    for lyric in lyrics_dict:
        if lyrics_dict[lyric] == highest_freq:
            words_list.append(lyric)
    return words_list, highest_freq
```

```python
most_common_words(lyrics_to_frequencies(beatles))
```

```python
most_common_words(lyrics_to_frequencies(yerin))
```

##3. words_often(lyrics_dict, minTimes)

image.png

```python
def words_often(lyrics_dict, minTimes):
    answer = []

    while True:
        if max(lyrics_dict.values()) >= minTimes:
            sorted_dict = sorted(lyrics_dict.items(), key=lambda x: x[1])
            word, count = sorted_dict.pop()
            answer.append(([word], count))
            del lyrics_dict[word]
        else:
            break

    return answer
```

```python
words_often(lyrics_to_frequencies(beatles), 6)
```

```python
words_often(lyrics_to_frequencies(yerin), 6)
```
