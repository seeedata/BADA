```python
## 판다스 임포트 하기
import pandas as pd
```

```python
position = ['president', 'vice-president', 'session manager', 'boss']
bada_2nd = ['Hoon', 'Hyun', 'Woo', 'Jong', 'Min', 'Kwon', 'Jeong', 'Soo', 'Park', 'Choi']
```

```python
## position과 bada_2nd가 length가 다른데 각 list의 길이 차이를 이용하여 position 에 그 차이만큼 'member' 추가하고 변수명(position2)에 저장하기
while True:
    position.append('member')
    if len(position) - len(bada_2nd) == 0:
        break
position2 = position
print(position2)
```

```python
department = ['Business', 'Business', 'Business', 'Business', 'Mathematics', 'Economics', 'Chinese', 'Sociology', 'Business', 'Business']
```

```python
## 이름을 index, 역할을 value로 하는 Series 만들어보기
series1 = pd.Series(position2, index = bada_2nd)
series1
```

```python
## 이름을 index, 학과을 value로 하는 Series 만들어보기
series2 = pd.Series(department, index = bada_2nd)
series2
```

```python
## pd.DataFrame 이용하여 각 칼럼명을 순서대로 'name', 'position', 'department' 로 하는 (인원,3) 모양의 df 구성하기(변수명 dict_df1에 저장할 것)
dict_df1 = pd.DataFrame({'name': bada_2nd,
                         'position': position2,
                         'department': department})
dict_df1
```

```python
## 위에 만든 df를 (3,인원수) 형태의 df로 바꾸고 변수명(dict_df2)에 저장하기
dict_df2 = dict_df1.T
dict_df2
```

```python
## dict_df2에 'name' 행을 이 df의 column명으로 바꾸기
dict_df2.columns = dict_df2.iloc[0]
dict_df2
```

```python
## iloc 이용하여 'name'행이 삭제되어 'position'과 'department' 두 행만이 존재하는 (2,인원수)의 df로 변환하고 변수명(dict_df3)에 저장하기'
dict_df3 = dict_df2.iloc[1:]
dict_df3
```

```python
## loc 이용하여 'name'행이 삭제되어 'position'과 'department' 두 행만이 존재하는 (2,인원수)의 df로 변환하고 변수명(dict_df4)에 저장하기
dict_df4 = dict_df2.loc['position':]
dict_df4
```

```python
## dict_df4의 차원확인하기
dict_df4.ndim
```

```python
## dict_df4의 형태(2,인원수) 확인하기
dict_df4.shape
```

```python
## department에서 중복값을 모두 제거하여 새로운 변수명(department2에 저장하기(*주의 : department2의 자료형이 list이어야 함)
department2 = list(set(department))
department2
```

```python
## position2에서 중복값을 모두 제거하여 새로운 변수명(position3에 저장하기(*주의 : position3의 자료형이 list이어야 함)
position3 = list(set(position2))
position3
```

```python
## 넘파이 임포트 하기
import numpy as np
```

```python
## 모든 값이 0 인 2x(max(department2, position3))의 배열생성하여 변수명(zero_mat)에 저장하기
a = max(len(department2), len(position3))
zero_mat = np.zeros((2, a))
print(zero_mat)
```

```python
## iloc, while, for 문을 사용하여, dict_df4에 있는 각 역할과 학과의 개수를 카운트 하고 position3와 department2에 저장된 순서에 따라 그 개수를 zero_mat에 표시하기
## 예시) department2가 ['사회학과', '중어중문학과', '식품자원경제학과', '경영학과', '수학과'] 라면
## zero_mat의 두번째 배열은 [1,1,1,6,1]이 되도록 한다.
## 다중 중첩문을 사용하여야 할 것으로 생각된다.

row = 0

while row < 2:

    col = 0

    while col < len(position3):
        if row == 0:
            zero_mat[row][col] = dict_df4.iloc[row].value_counts()[position3[col]]
        elif row == 1:
            zero_mat[row][col] = dict_df4.iloc[row].value_counts()[department2[col]]

        col += 1

    row += 1

print(zero_mat)
```

```python
## zero_mat이 원하는대로 잘 저장되었는지 출력해보자
zero_mat
```

```python
## 변수명(position_population)에 zero_mat안의 각 역할의 개수를 저장한 배열을 value로 하고 position3을 index로 하는 Series를 만들어 저장하여라
position_population = pd.Series(zero_mat[0], index = position3)
print(position_population)
```

```python
## 변수명(department_population)에 zero_mat안의 각 학과의 개수를 저장한 배열을 value로 하고 department2을 index로 하는 Series를 만들어 저장하여라
department_population = pd.Series(zero_mat[1], index = department2)
print(department_population)
```
