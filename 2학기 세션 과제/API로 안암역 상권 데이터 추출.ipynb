동일 Open API에서 상권 내 상가업소 조회 API를 활용해 안암역 근처의 상권에 해당하는 데이터들을 추출하세요.

```python
import requests
import json
import pandas as pd
```

```python
#url 입력
address = 'https://apis.data.go.kr/B553077/api/open/sdsc2/storeListInArea'
ServiceKey = '?ServiceKey=m7eM%2FA2PGstpvx63aua%2Fj%2BpjRsDXnGWhz9hDKDh0ftnzjLXh8hYZuP6zHitcRQlZ4aiSbMvTsWYviUZ5JmOsuQ%3D%3D'
pageNo = 1
numOfRows = 10
key = 9392

url = address + ServiceKey + '&pageNo=' + str(pageNo) + '&numOfRows=' + str(numOfRows) + '&key=' + str(key) + '&type=json'

print(url)
```

```python
#url 불러오기
response = requests.get(url)

#데이터 값 출력해보기
contents = response.text

#데이터 출력
contents
```

```python
#데이터 형식 변환
contents = json.loads(contents)
print(type(contents))
contents
```

```python
col = contents['header']['columns']
df = pd.DataFrame(columns=col)

for i in range(10**10):
    try:
        new_row_df = pd.DataFrame(list(contents['body']['items'][i].values())).transpose()
        new_row_df.columns = col
        df = pd.concat([df, new_row_df], ignore_index=True)
    except:
        break

df
```
