```python
def get_months_in_range(date_range):
    start_date, end_date = date_range.split(' ~ ')
    start_year, start_month, _ = map(int, start_date.split('.'))
    end_year, end_month, _ = map(int, end_date.split('.'))

    months = []
    current_year = start_year
    current_month = start_month

    while current_year < end_year or (current_year == end_year and current_month <= end_month):
        month_name = str(current_month) + '월'
        months.append(month_name)

        current_month += 1
        if current_month > 12:
            current_month = 1
            current_year += 1

    return months

def get_months_between_dates(start_date_str, end_date_str):
    # 날짜 형식 지정
    date_format = "%Y-%m-%d"

    # 시작 날짜와 종료 날짜를 파싱
    start_date = datetime.strptime(start_date_str, date_format)
    end_date = datetime.strptime(end_date_str, date_format)

    # 시작 날짜의 월부터 종료 날짜의 월까지의 범위 내의 달을 구함
    months_between = []
    current_month = start_date.month
    current_year = start_date.year

    while (current_year, current_month) <= (end_date.year, end_date.month):
        months_between.append(f"{current_month}월")

        # 다음 달로 이동
        current_month += 1
        if current_month > 12:
            current_month = 1
            current_year += 1

    return ' '.join(months_between)


def make_season(data):
  # 각 계절에 해당하는 월을 나타내는 딕셔너리
  seasons = {
      '겨울': ['12월', '1월', '2월'],
      '봄': ['3월', '4월', '5월'],
      '여름': ['6월', '7월', '8월'],
      '가을': ['9월', '10월', '11월']
  }

  # 데이터를 월로 분할
  months = data.split(', ')

  # 각 계절을 판별하고 텍스트 추가
  result = ''
  for season, months_list in seasons.items():
      if any(month in months_list for month in months):
          result += season + ' '

  # 결과 출력
  return(result.strip())
```

## 모델 제작

```python
max_epochs = 60

model = Doc2Vec(
    dm_concat=1,
    window=15,
    vector_size=300,
    alpha=0.025,
    min_alpha=0.0001,
    min_count=2,
    dm=0,
    negative=5,
    seed=42,
    workers=8)

model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    # decrease the learning rate
    model.alpha = model.alpha*0.97
    print(model.alpha)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def similar_vector_index(index,n):

  target_vector = get_sentence_vector(df2['텍스트'].iloc[index])

  df3 = df2[df2['RENTLSPCE_END_DE'] < df2['RENTLSPCE_BEGIN_DE'][index]]

  Vector = []
  for i in range(len(df3)):
     Vector.append(get_sentence_vector(df3['텍스트'].iloc[i]))

  similarities = cosine_similarity([target_vector], Vector)[0]

  most_similar_indices = np.argsort(similarities)[::-1][:n]

  return(df3.iloc[most_similar_indices])
```

```python
def similar_vector_sentence(sentence, date ,n):

  target_vector = get_sentence_vector(sentence)

  df3 = df2[df2['RENTLSPCE_END_DE'] < date]

  Vector = []
  for i in range(len(df3)):
     Vector.append(get_sentence_vector(df3['텍스트'].iloc[i]))

  similarities = cosine_similarity([target_vector], Vector)[0]

  most_similar_indices = np.argsort(similarities)[::-1][:n]

  return(df3.iloc[most_similar_indices])
```
