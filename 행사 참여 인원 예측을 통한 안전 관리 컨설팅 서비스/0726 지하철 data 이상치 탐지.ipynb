#표준화 잔차 만들기

```python
def standardized_residuals(df):
    for column in ['resid_00~01', 'resid_01~02', 'resid_02~03', 'resid_04~05',
       'resid_05~06', 'resid_06~07', 'resid_07~08', 'resid_08~09',
       'resid_09~10', 'resid_10~11', 'resid_11~12', 'resid_12~13',
       'resid_13~14', 'resid_14~15', 'resid_15~16', 'resid_16~17',
       'resid_17~18', 'resid_18~19', 'resid_19~20', 'resid_20~21',
       'resid_21~22', 'resid_22~23', 'resid_23~24']:
        new_col_name = 'std{}'.format(column)
        df[new_col_name] = (df[column] - np.mean(df[column])) / np.std(df[column])
    return df.head()
```
