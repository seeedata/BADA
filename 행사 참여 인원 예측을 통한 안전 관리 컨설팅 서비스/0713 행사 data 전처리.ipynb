##### Column 설명

FCLTY_NM: 시설명

EVENT_SDIV_NM: 행사 구분명

EVENT_NM: 행사명

RENTLSPCE_BEGIN_DE: 대관시작일자

RENTLSPCE_END_DE: 대관종료일자

USE_DAY_CO: 사용일수

REQST_EVENT_NMPR_CO: 신청행사인원수

EXCCLC_EVENT_NMPR_CO: 정산행사인원수

ISE_ELSE_FLAG_NM: 내외구분명

```python
df = df.drop(df.columns[[0,1,3,11]], axis=1)
```

1. 날짜정렬

- 행사 시작 시점으로 오름차순 정렬

```python
df = df.sort_values("RENTLSPCE_BEGIN_DE")
df
```

2. 신청, 정산인원 둘 다 결측치이거나 0인 행 drop 하기

```python
df.isnull().sum()
```

```python
# 공연구분 정의 안된 행 drop
df = df[~ df["EVENT_SDIV_NM"].isnull()]
```

```python
from datetime import datetime
```

```python
df["RENTLSPCE_BEGIN_DE"]= df["RENTLSPCE_BEGIN_DE"].astype('str')
```

```python
df["RENTLSPCE_END_DE"]= df["RENTLSPCE_END_DE"].astype('str')
```

```python
df["RENTLSPCE_BEGIN_DE"] = pd.to_datetime(df["RENTLSPCE_BEGIN_DE"])
```

```python
df["RENTLSPCE_END_DE"] = pd.to_datetime(df["RENTLSPCE_END_DE"])
```

```python
# 사용일수 계산 (종료일자-시작일자 +1)
diff = df["RENTLSPCE_END_DE"]- df["RENTLSPCE_BEGIN_DE"]
df["USE_DAY_CO"] = diff
df
```

```python
len(df.dropna(subset=["REQST_EVENT_NMPR_CO", "EXCCLC_EVENT_NMPR_CO"]))
```

```python
len(df[(df["REQST_EVENT_NMPR_CO"] != 0) | (df["EXCCLC_EVENT_NMPR_CO"] != 0)])
```

```python
# 신청인원, 정산인원 결측치인 행 제거
df = df.dropna(subset=["REQST_EVENT_NMPR_CO", "EXCCLC_EVENT_NMPR_CO"])
```

```python
# 신청인원, 정산인원 0인 행 제거
df = df[(df["REQST_EVENT_NMPR_CO"] != 0) | (df["EXCCLC_EVENT_NMPR_CO"] != 0)]
```

```python
df.shape # 495개의 행
```

```python
#인덱스 재정렬
df.reset_index(drop=True, inplace =True)
```

3. 훈련 등 관련없는 word 제외

```python
index = df[df["EVENT_NM"].str.contains("훈련")].index
```

```python
df.drop(index, inplace=True)
```
