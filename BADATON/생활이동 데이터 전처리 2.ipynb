```python
df=pd.read_csv('/content/drive/MyDrive/BADA/바다톤/data/생활이동데이터_일부전처리.csv')
df.head()
```

```python
#나이를 주어진 범주로 묶는 함수
def calculate_age_group(age):
    if age >= 20 and age < 30:
        return '20대'
    elif age >= 30 and age < 40:
        return '30대'
    elif age >= 40 and age < 50:
        return '40대'
    elif age >= 50 and age < 60:
        return '50대'
    elif age >= 60 and age < 70:
        return '60대'
    elif age >= 70 and age <80:
        return '70대'
    elif age >= 80 and age <90:
        return '80대'

#나이대 계산 및 데이터프레임에 추가
df['나이대'] = df['나이'].apply(calculate_age_group)

#나이대별로 그룹화하여 이동인구(합) 계산
df1 = df.groupby(['대상연월', '요일', '도착시간', '출발 시군구 코드', '도착 시군구 코드', '성별', '나이대','이동유형'])['이동인구(합)'].sum().reset_index()

df1
```

```python
df1['나이대'].value_counts()
```

2023.02 21시데이터는 20시로, 22시 데이터는 23시로 채우기

```python
df220 = df1.loc[(df1['대상연월'].isin(['2023-02-01'])) & (df1['도착시간'] == 20)]
df220
```

```python
df221 = df220.replace({'도착시간': 20}, 21)
df221
```

```python
df223= df1.loc[(df1['대상연월'].isin(['2023-02-01'])) & (df1['도착시간'] == 23)]
df223
```

```python
df222 = df223.replace({'도착시간': 23}, 22)
df222
```

```python
df2 = pd.concat([df1, df221], ignore_index=True)
df2
```

```python
df3 = pd.concat([df2, df222], ignore_index=True)
df3
```

```python
df3.loc[(df3['대상연월'].isin(['2023-02-01'])) & (df3['도착시간'] == 22)]
```

```python
df4 = df3.sort_values(by=['대상연월','도착시간'], ascending=[True, True])
df4
```

```python
#도착시간 추출 전 문자열로 바꿔주기
df4['도착시간']=df4['도착시간'].astype(str)
df4.dtypes
```

```python
#시간 데이터 변환 함수
def convert_to_time(hour):
    return f'{hour.zfill(2)}:00'  #hour.zfill(2): hour 값을 두 자리 숫자로 만든다.
                                  #이후 :00을 붙여 시간 형식으로 만든다.

df4['도착시간'] = df4['도착시간'].apply(convert_to_time)
```

```python
df4['도착시간'].replace('00:00', '0:00', inplace=True)
```

```python
df4
```

```python
#errors='raise': 형식 변환 중 오류가 발생하면 오류를 발생
#dt.time을 사용해 날짜 부분 제외하고 시간 부분만 유지
df4['도착시간'] = pd.to_datetime(df4['도착시간'], format='%H:%M', errors='raise').dt.time
```

```python
df5 = df4.rename(columns={'도착시간': '귀가시간대', '이동인구(합)': '이동인구수'})
df5
```

```python
df5['출발 시군구 코드'] = df5['출발 시군구 코드'].astype(str)
df5['도착 시군구 코드'] = df5['도착 시군구 코드'].astype(str)
df5
```
